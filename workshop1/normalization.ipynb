```python
import dlt
import duckdb
import requests
```

```python
def load_plp_data():
    url = 'https://api.digikala.com/v1/promotions/plp_99533388/'
    response = requests.get(url)
    response.raise_for_status()
    page_json = response.json()
    yield page_json['data']
```

```python
pipeline = dlt.pipeline(destination="duckdb", dataset_name="dk", credentials="data.db")
```

```python
info = pipeline.run(load_plp_data, table_name="plp", write_disposition="replace")
print(info)
```

```python
print(pipeline.dataset_name)
conn = duckdb.connect("data.db")
conn.sql(f"SET search_path = '{pipeline.dataset_name}'")
tables = conn.sql("show tables").df()
```
